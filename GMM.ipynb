```python
df=pd.read_csv("/kaggle/input/covid-19-prediction/master_dataset.csv",low_memory=False)
```

```python
current_columns = df.columns.tolist()
new_columns = current_columns[1:] + ['empty_column']
df_fixed = df.copy()
df_fixed.columns = new_columns
df = df_fixed.drop('empty_column', axis=1)
```

```python
df = df[['sex', 'age', 'bmi','smoking','alcohol','cannabis','amphetamines','cocaine','contacts_count','working','rate_reducing_risk_single','rate_reducing_mask','covid19_symptoms','covid19_contact','asthma','kidney_disease','liver_disease','compromised_immune','heart_disease','lung_disease','diabetes','hiv_positive','hypertension','other_chronic','nursing_home','health_worker','covid19_positive']]
```

```python
def process_age_column(df, age_column='age'):
    """
    Process age column containing string ranges like '20_30' to their average values
    and handle null values by replacing them with mean.
    
    Parameters:
    df: pandas DataFrame
    age_column: str, name of the age column (default='age')
    
    Returns:
    pandas Series with processed age values
    """
    def extract_average(age_str):
        if pd.isna(age_str):
            return np.nan
        try:
            # Split the string on '_' and convert to integers
            start, end = map(int, str(age_str).split('_'))
            return (start + end) / 2
        except:
            return np.nan
    
    # Convert string ranges to averages
    processed_ages = df[age_column].apply(extract_average)
    
    # Replace null values with mean
    mean_age = processed_ages.mean()
    processed_ages = processed_ages.fillna(mean_age)
    
    return processed_ages

df['age'] = process_age_column(df, age_column='age')
```

```python
categorical_cols = ['sex', 'age', 'smoking', 'alcohol', 'working', 'cannabis', 
                       'amphetamines', 'cocaine']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
numerical_cols = ['bmi', 'contacts_count', 'rate_reducing_mask']
for col in numerical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
```

```python
nominal_cols = ['sex', 'smoking', 'working']
df_encoded = pd.get_dummies(df, columns=nominal_cols)
drug_cols = ['cannabis', 'amphetamines', 'cocaine']
for col in drug_cols:
    if col in df_encoded.columns:
        df_encoded[col] = pd.to_numeric(df_encoded[col], errors='coerce')
df=df_encoded
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler  # you can also try MinMaxScaler
from sklearn.metrics import (classification_report, accuracy_score, precision_score, 
                             recall_score, f1_score, confusion_matrix, roc_curve, auc)
from imblearn.combine import SMOTEENN
from sklearn.linear_model import LogisticRegression

# ---------------------------
# 1. Data Preprocessing
# ---------------------------

# Suppose 'df' is your initial DataFrame

# Convert boolean columns to int (True becomes 1, False becomes 0)
bool_cols = df.select_dtypes(include=['bool']).columns
df[bool_cols] = df[bool_cols].astype(int)

# Function to remove outliers using IQR method
def remove_outliers(df, features, multiplier=1.5):
    for col in features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Exclude target column from features list
cols_to_check = [col for col in df.columns if col != 'covid19_positive']
df = remove_outliers(df, cols_to_check)

# ---------------------------
# 2. Scaling and Resampling
# ---------------------------

# Normalize features using StandardScaler
scaler = StandardScaler()
cols_to_normalize = [col for col in df.columns if col != 'covid19_positive']
df_normalized = df.copy()
df_normalized[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])

# Apply SMOTEENN for balancing classes
X_features = df_normalized[cols_to_normalize]
y_target = df_normalized['covid19_positive']

smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_features, y_target)

# Convert back to DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=cols_to_normalize)
df_resampled['covid19_positive'] = y_resampled

# Use the resampled DataFrame for modeling
df_final = df_resampled

# ---------------------------
# 3. Train-Test Split
# ---------------------------
X = df_final.drop('covid19_positive', axis=1).values
y = df_final['covid19_positive'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# ---------------------------
# 4. GMM Classifier
# ---------------------------
# Tune GMM: using 2 components per class and 'diag' covariance for simplicity.
gmm0 = GaussianMixture(n_components=2, covariance_type='diag', random_state=42)
gmm1 = GaussianMixture(n_components=2, covariance_type='diag', random_state=42)

# Fit GMM on training data for each class separately
gmm0.fit(X_train[y_train == 0])
gmm1.fit(X_train[y_train == 1])

# Compute log-likelihoods for test samples for each class
ll0 = gmm0.score_samples(X_test)
ll1 = gmm1.score_samples(X_test)

# Convert log-likelihoods to probabilities using softmax
max_ll = np.maximum(ll0, ll1)
prob0 = np.exp(ll0 - max_ll)
prob1 = np.exp(ll1 - max_ll)
sum_prob = prob0 + prob1
p0 = prob0 / sum_prob
p1 = prob1 / sum_prob

# ---------------------------
# 5. Optimize Threshold using ROC (Youden's J statistic)
# ---------------------------
fpr, tpr, thresholds = roc_curve(y_test, p1)
J = tpr - fpr  # Youden's J statistic
optimal_idx = np.argmax(J)
optimal_threshold = thresholds[optimal_idx]
print("Optimal threshold:", optimal_threshold)

# Predict using the optimal threshold
y_pred_gmm = (p1 > optimal_threshold).astype(int)

# ---------------------------
# 6. Evaluation of GMM Classifier
# ---------------------------
accuracy = accuracy_score(y_test, y_pred_gmm)
precision = precision_score(y_test, y_pred_gmm, zero_division=0)
recall = recall_score(y_test, y_pred_gmm, zero_division=0)
f1 = f1_score(y_test, y_pred_gmm, zero_division=0)
cm = confusion_matrix(y_test, y_pred_gmm)

print("\n=== Gaussian Mixture Model Classifier ===")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nClassification Report:\n", classification_report(y_test, y_pred_gmm, zero_division=0))
print("\nConfusion Matrix:\n", cm)

# Plot ROC Curve for GMM classifier
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - GMM Classifier')
plt.legend(loc="lower right")
plt.show()

# ---------------------------
# 7. Optional: Compare with Logistic Regression
# ---------------------------
logreg = LogisticRegression(random_state=42, max_iter=1000)
logreg.fit(X_train, y_train)
y_pred_lr = logreg.predict(X_test)

print("\n=== Logistic Regression Classifier ===")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr, zero_division=0))
print("Recall:", recall_score(y_test, y_pred_lr, zero_division=0))
print("F1 Score:", f1_score(y_test, y_pred_lr, zero_division=0))
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr, zero_division=0))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
```
